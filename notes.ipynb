# Derivadas

La derivada mide la tasa de cambio de una función respecto a su variable independiente. En otras palabras: la derivada te dice en qué proporción cambia la salida o output de una función cuando el dato de entrda incrementa o decrementa

$$
\begin{align}
\frac{df}{dx} &= \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x} \\
\end{align}
$$

## Ejemplo

Dada la siguiente función:

$$
f(x) = x^2
$$

Su derivada es:

$$
\begin{aligned}
\frac{df}{dx}
&= \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x}\\
&= \lim_{\Delta x \to 0} \frac{(x + \Delta x)^2 - (x)^2}{\Delta x}\\
&= \lim_{\Delta x \to 0} \frac{(x^2 + 2x \cdot \Delta x + \Delta x^2) - (x)^2}{\Delta x}\\
&= \lim_{\Delta x \to 0} \frac{(\cancel{x^2} + 2x \cdot \Delta x + \Delta x^2) - \cancel{(x)^2}}{\Delta x}\\
&= \lim_{\Delta x \to 0} \frac{2x \cdot \Delta x + \Delta x^2}{\Delta x}\\
&= \lim_{\Delta x \to 0} \frac{\cancel{\Delta x} \cdot (2x  + \Delta x)}{\cancel{\Delta x}}\\
&= \lim_{\Delta x \to 0} 2x  + \Delta x\\
&= 2x  + (0) \\
&= \boxed{2x}
\end{aligned}
$$

## Ejemplo en Python

**Definición de la función en Python**

```python
def f(x):
  return x**2
```

**Gráfica de la función entre -5 y 5**

```python
start, end, step = -5, 5, 0.25

X = np.arange(start, end, step)
Y = f(X)

plt.plot(X, Y)
plt.show()
```

```python
h = 0.000001

x = 4   # Play with this and notice how the derivative explains the increment when x changes
y = (f(x + h) - f(x)) / h

print(y)
```

```python
def plot_tangent_line(point_x, slope):
    plt.plot(X, Y, label="f(x)")
    
    # Mark the point of tangency
    tangent_point_x = point_x
    tangent_point_y = f(point_x)

    plt.scatter([tangent_point_x], [tangent_point_y], color='green')
    
    # Tangent line calculation for a range around x
    def tangent_line(tangent_line_X):
        return slope * (tangent_line_X - point_x) + tangent_point_y
        
    # Plot red line representing the tanget line at point x
    tangent_line_X = np.arange(tangent_point_x-1, tangent_point_x+1, 0.25)
    tangent_line_Y = tangent_line(tangent_line_X)

    plt.plot(tangent_line_X, tangent_line_Y, color='red')
```

```python
def df(x):
  return 2*x
```

```python
x = 4
derivative = df(x)

plot_tangent_line(x, derivative)

print(derivative)
```

## Ejemplo más complejo (derivadas parciales)

```python
def g(a, b, c):
  return a * b + c
```

**¿Qué tanto afecta el cambio en una de las variables al resultado final?**

```python
h = 0.0001

g1 = g(2, -3, 10)

print('g1:', g1)

g2 = g(2, -3, 10 + h) # Play with h on each parameter
print('g2:', g2)

print('pendiente:', (g2 - g1) / h) 
```
